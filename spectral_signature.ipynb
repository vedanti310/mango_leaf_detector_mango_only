```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data
df = pd.read_csv("mango_rgb_data.csv")  # Change to your actual file name

# Step 2: Filter mango leaf data (Label = 1)
mango_df = df[df['Label'] == 1]

# Step 3: Compute average RGB values
mean_r = mango_df['R'].mean()
mean_g = mango_df['G'].mean()
mean_b = mango_df['B'].mean()

# Step 4: Plot RGB spectral signature
plt.figure(figsize=(6, 4))
plt.plot(['Red', 'Green', 'Blue'], [mean_r, mean_g, mean_b],
         marker='o', linestyle='-', color='green', linewidth=2)

# Optional: Add value labels
for i, val in enumerate([mean_r, mean_g, mean_b]):
    plt.text(i, val + 1, f"{val:.1f}", ha='center')

# Styling
plt.title("Spectral Signature of Mango Leaf (RGB)")
plt.xlabel("Color Bands")
plt.ylabel("Average Reflectance (Pixel Intensity)")
plt.ylim(0, 200)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv("mango_rgb_data.csv")

# Step 2: Filter only mango leaf (Label = 1)
mango_df = df[df["Label"] == 1]

# Step 3: Compute mean RGB values
mean_r = mango_df["R"].mean()
mean_g = mango_df["G"].mean()
mean_b = mango_df["B"].mean()

# Step 4: Normalize to get reflectance
reflectance_r = mean_r / 255
reflectance_g = mean_g / 255
reflectance_b = mean_b / 255

# Step 5: Wavelength mapping
wavelengths = [470, 550, 640]  # Blue, Green, Red
reflectance = [reflectance_b, reflectance_g, reflectance_r]

# Step 6: Plot reflectance vs. wavelength
plt.figure(figsize=(6, 4))
plt.plot(wavelengths, reflectance, marker='o', linestyle='-', color='purple')
plt.title("Spectral Signature of Mango Leaf (Reflectance vs Wavelength)")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Reflectance (Normalized RGB)")
plt.ylim(0, 1)
plt.grid(True)

# Optional: add labels on points
for i, val in enumerate(reflectance):
    plt.text(wavelengths[i], val + 0.02, f"{val:.2f}", ha='center')

plt.tight_layout()
plt.show()
```
